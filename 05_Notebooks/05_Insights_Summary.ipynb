# Beyond Last-Click: Attribution Analysis
## Insights Summary — Soheila Zamani | SPICED Academy Berlin | April 2026

This notebook summarizes all key findings from the YouTube Influencer Analysis and A/B Testing datasets.
It covers 6 hypotheses (4 core + 2 bonus) across two datasets: 3,848 YouTube creators (after filtering subscribers ≥ 1,000) and 588,101 A/B test users.

## H1 | Nano & Micro Influencers Drive Higher Engagement Per Subscriber

**Finding:** Nano YouTube creators (< 10K subscribers) achieve a median view-per-subscriber ratio of 0.884,
compared to 0.326 for mega YouTube creators (> 1M subscribers). That is approximately 2.7× more engagement.

**Interpretation:** Smaller YouTube creators have more loyal, niche audiences. Their content feels personal
and authentic, which drives higher relative viewership per follower.

**Business implication:** Brands seeking high engagement-per-dollar should prioritize nano and micro
YouTube creators over mega creators for influencer marketing campaigns.

## H2 | Fashion, Lifestyle & Beauty Lead Engagement Across All Niches

**Finding:** Fashion leads with median VPS 0.71. Fitness shows the lowest engagement. Visual and aspirational niches consistently outperform other categories.

**Interpretation:** Visual and aspirational niches attract highly engaged viewers who return regularly.
These audiences are also commercially valuable for e-commerce brands.

**Business implication:** E-commerce brands in lifestyle categories should focus influencer budgets
on Fashion, Lifestyle, and Beauty YouTube creators regardless of tier.

## H3 | Regression: Creator Size Negatively Predicts Engagement Rate

**Method:** Log-linear regression — log10(subscribers) → log10(view_per_subscriber)

**Results:**
- R² = 0.27 → creator size explains ~27% of the variation in engagement rate
- Coefficient = −0.23 → each 10× increase in subscribers reduces VPS by ~41%
- Spearman r = −0.51, p-value < 0.001 → statistically significant negative correlation

**Interpretation:** As YouTube creators grow in subscriber count, their engagement rate reliably declines.
The relationship is moderate — other factors (content quality, niche, upload frequency) also matter.

**Limitation:** R² of 0.27 means 73% of engagement variation is unexplained by size alone.

## H4 | Paid Ads Convert 43% Better Than Organic (A/B Test)

**Dataset:** 588,101 users — split into ad group (paid) and organic group

**Results:**
- Organic conversion rate: 1.79%
- Paid ads conversion rate: 2.55%
- Conversion lift: (2.55 − 1.79) / 1.79 × 100 = **+42.5% lift**

**Interpretation:** Users exposed to paid ads are 43% more likely to convert than users who
browse organically. This is a statistically meaningful lift at this sample size.

**Business implication:** Paid ad spend produces measurable ROI above organic baseline.
The investment in ad campaigns is justified by conversion uplift.

## H5 (Bonus) | Geography & Content Volume Reveal Hidden Patterns

**Geographic finding:**
Among identified countries with a minimum of 10 YouTube creators, Brazil (VPS 1.12), Turkey (0.79), and South Africa (0.75) show the highest median engagement per subscriber.
All three are dominated by nano/micro YouTube creators in Fashion, Lifestyle, and Beauty niches —
directly reflecting H1 and H2 findings at a country level.

**Note on Unknown country group:**
925 YouTube creators (20% of the dataset) have no country recorded. This group shows the highest
median VPS of all (1.59). Based on creator names, the majority appear to be South Asian creators
(likely India-based) in Beauty and Fashion niches who did not set a country in their YouTube profile.
Their high engagement is consistent with H1 — they are 44% nano creators in top-performing niches.
Excluding this group does not change the top 3 identified country findings (Brazil, Turkey, South Africa).

**Content volume finding:**
YouTube creators with higher video counts show lower views-per-subscriber (negative trend, log-log scale).
Higher output volume correlates with lower per-video engagement — consistent with H3.

**Limitation:** YouTube API keyword search may overrepresent established niches in certain countries.
Results reflect available data, not the full global YouTube landscape.

## H6 (Bonus) | Ad Timing Matters — Saturday Mornings and Weekday Afternoons Convert Best

**Method:** Heatmap of paid conversion rate by day of week × hour of day (7 × 24 grid)

**Key findings:**
- Highest conversion: Saturday 5:00–6:00 (up to 5.68%) — the single strongest window
- Strong weekday performers: Tuesday 16:00 (4.56%), Monday 14:00–16:00 (~4.3–4.5%)
- Wednesday 16:00 also performs well (3.57%) — consistent with mid-week afternoon pattern
- Weakest windows: Early morning hours 0–4 across most weekdays

**Pattern:** Two distinct peaks emerge — Saturday early morning and weekday early-to-mid afternoon
(14:00–16:00). These likely reflect different audience behaviors: weekend leisure browsing
and weekday lunch-break or post-work browsing.

**Business implication:** Prioritize Saturday 5–7am and Tuesday/Monday 14–16:00 for
highest-ROI ad scheduling. Wednesday 16:00 is a strong secondary window.
Avoid overnight weekday scheduling for minimum waste.

## Data Limitations

1. **YouTube API sampling bias:** The dataset (3,848 YouTube creators after filtering — subscribers ≥ 1,000) was collected via
   keyword search. Nano YouTube creators (474) are underrepresented vs. macro (1,339) and mega (1,070), which is the inverse
   of a typical YouTube distribution. This may overestimate engagement patterns typical for smaller creators.
   Future analysis should rebalance sampling to better reflect real platform distribution.

2. **Engagement proxy:** View-per-subscriber is used as the engagement metric. It does not capture
   likes, comments, shares, or click-through rates. Future analysis should include these signals
   from the YouTube API for a more complete engagement view.

3. **A/B test generalizability:** The A/B dataset does not specify the product, industry, or ad creative.
   The +43% lift is real but results should be validated within each specific business context
   before scaling ad budgets.

4. **Geographic coverage:** The country filter (min. 10 YouTube creators per country) excludes
   smaller markets, which may bias insights toward larger markets and miss emerging opportunities
   in high-growth regions.

5. **Unknown country group:** 925 YouTube creators (20%) have no country data recorded.
   This group ranks #1 by engagement (median VPS = 1.59) and likely consists of South Asian
   nano creators in Beauty and Fashion. Excluding them does not change the top 3 identified country findings (Brazil, Turkey, South Africa), but their origin remains unconfirmed.

## From Limitations to Business Decisions

Each limitation has a direct implication for how findings should be applied in practice:

| Limitation | Business implication |
|---|---|
| Nano YouTube creators overrepresented | Engagement benchmarks may be optimistic — validate with your own creator data before setting KPIs |
| Engagement proxy (VPS only) | Do not select YouTube creators on VPS alone — add comment rate and CTR to the evaluation criteria |
| A/B test context not specified | Run your own A/B test with your product before assuming +43% lift applies to your campaign |
| Country filter excludes small markets | Brazil/Turkey/South Africa findings are solid — but high-growth opportunities may exist in smaller markets |
| Unknown country group = highest VPS | Worth investigating — identifying these creators could reveal an untapped high-engagement market |

## Conclusions & Recommendations

Invest in nano and micro YouTube creators in Fashion, Lifestyle, and Beauty.
They deliver 2.7× more engagement per subscriber at a fraction of the cost of mega creators.

**For paid advertising:**
Paid ads generate a 43% conversion lift over organic. Budget allocation should favor paid advertising,
with scheduling weighted toward Saturday 5–7am and Monday/Tuesday 14–16:00.

**For market targeting:**
Brazil, Turkey, and South Africa are the top identified high-engagement markets, dominated by
nano/micro YouTube creators. These are strong candidates for early-stage or growth-phase campaigns.
A large group of unidentified creators (likely South Asian) also shows very high engagement

**Overall:**
The data consistently shows that smaller, more targeted YouTube creators combined with
well-timed paid ads outperform broad, high-volume strategies in e-commerce conversion contexts.
The brands that will win are the ones who go beyond the last click — and understand the full picture.

## Bonus Finding — Enrichment: Engagement Rate vs. Reach (VPS)

YouTube API enrichment (likes + comments per video, n=1,000 sampled channels) revealed 
that **engagement rate and VPS measure fundamentally different things** (correlation: −0.1).

| Tier  | VPS (H1) | Engagement Rate |
|-------|----------|-----------------|
| Nano  | 0.884 ✅ | 0.0233 ❌ lowest |
| Micro | 0.731    | 0.0305           |
| Macro | 0.461    | 0.0354 ✅ highest |
| Mega  | 0.326    | 0.0316           |

**Interpretation:** Nano creators drive more views per subscriber (reach/awareness).  
Macro creators drive more likes and comments per view (interaction/conversion intent).

**By niche:** Film & Streaming (0.0383) and Food (0.0370) lead on engagement,  
while Fashion (0.0233) leads on VPS — opposite rankings.

**Strategic implication:** Choose nano/Fashion for awareness campaigns.  
Choose macro/Film & Food for engagement and conversion campaigns.
